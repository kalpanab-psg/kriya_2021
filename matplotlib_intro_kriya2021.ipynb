```python
# multiple variables in a single plot
#storing minimum, maximmum , average temperatures in lists

days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
#plot  min_t with day

plt.xlabel("day")
plt.ylabel("Minimum temp")
plt.plot(days,min_t,color='r')
```

```python
#plot all data in one chart
plt.plot(days,min_t)
plt.plot(days,max_t)
plt.plot(days, avg_t)
```

```python
#name the lines
plt.plot(days,min_t, label="min")
plt.plot(days,max_t,label="max")
plt.plot(days, avg_t,label="avg")

#label the axis and title
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather")
plt.legend()  # set legand property to show legand
```

```python
# set location for legend
plt.plot(days,min_t, label="min")
plt.plot(days,max_t,label="max")
plt.plot(days, avg_t,label="avg")

#label the axis and title
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather")
plt.legend(loc="upper right")  # set legand property to show legand

```

##  Applying pseudocolor schemes to image plots
Pseudocolor can be a useful tool for enhancing contrast and visualizing your data more easily. 
This is especially useful when making presentations of your data using projectors - their contrast is typically quite poor.
Pseudocolor is only relevant to single-channel, grayscale, luminosity images. We currently have an RGB image. 
Since R, G, and B are all similar (see for yourself above or in your data), we can just pick one channel of our data:
    

```python
lum_img = img[:, :, 0]

# This is array slicing.  You can read more in the `Numpy tutorial
# <https://docs.scipy.org/doc/numpy-dev/user/quickstart.html>`_.

plt.imshow(lum_img);
```

```python
plt.imshow(lum_img, cmap="hot");
```

```python
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('nipy_spectral')
```

Color scale reference
Its helpful to have an idea of what value a color represents. We can do that by adding color bars.

```python
imgplot = plt.imshow(lum_img)
plt.colorbar()
```
